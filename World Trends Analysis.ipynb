```python
# Visualisation
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
%matplotlib inline 
plt.rcParams['figure.figsize'] = 8,5
import warnings
warnings.filterwarnings('ignore')
```

```python
stats.head()
```

```python
# Distribution
vis1 = sns.distplot(stats['Internet users'], bins=30)
```

```python
vis2 = sns.boxplot(data=stats, x='Income Group', y="Birth rate")
```

Majoriy of the high income countries have a birth rate of 12 with a couple of outlies having both a high income and a high birth rate of 35.

Majority of low income countries have high birth rates of 37 with the first quartile ranging from 25-35, the second quartile ranging from 35-37 and the forth quartile ranging from 42-49.HOwever there is an ouliter having both low income levels and a low birth rate of 20.

Birth rates of Upper middle income countries range from 10-33, with majority being at 28 and an outlier being at 47.

Lower middle income countries have a birth rate rangine between 12-43 with and average birth rate of 26.

```python
# Birthrate to Internet users linear regression model
vis3 = sns.lmplot(data=stats, x='Internet users', y='Birth rate',
                  fit_reg =False, hue='Income Group', size=7, aspect=2,
                 scatter_kws={'s':100})
```

The graph illustrates lower birth rates corresponding to high internet usage levels indicating that on average, birth rates decerease as internet usage levels increase.

All countries with low income levels have high birth rates and low numbers of internet users. Lower middle income countries are fairly distributed from average to high rates of birth with their internet usage numbers topping out in the mid 40's with one outlier in the mid 50's. 

Upper middle income coiuntries  have birth rates ranging from low to medium levels, peaking at 30 with internet user naumber rangin from 15%-70% with most of the countries having internet user number s rangin from 25%-50%.

High income countries have low birth rates ranging from 5%-20%, with one outlier having a high birth rate of 35%. Their internet user numbers range from 42%-98% with one outlier, the one with a high birth rate having low a internet user level at 18%.
